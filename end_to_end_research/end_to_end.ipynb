### Check Data Distribution
There are a number of ways we can do this, however, one of the most common ways is through the use of a histogram.
We will use the pandas library to do this. We will also utilize a for loop so that we can iterate through each numerical variable. (Please see documentation here: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.hist.html)

First, we define a function that iterates through all the columns in our dataframe, and performs the pandas hist function to generate the plots.

```python
def plot_histograms(dataframe):
    for column in dataframe.columns:
        plt.hist(dataframe[column], bins=20)
        plt.title(f'Histogram for {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```

We then call the function and pass our dataframe as a parameter. This displays the histograms for each numerical feature.

```python
plot_histograms(df)
```
